```python
# Pivot
df_20 = pd.DataFrame({'Sr no.' : [1,2,3,4,5], 'A' : ['apple', 'ant', 'arm', 'air', 'aeroplane'], 'B' : ['banana', 'bat', 'ball', 'battery', 'bag'], 'C' : ['cat', 'camera', 'cape', 'cucumber', 'capital'], 'D' : ['dog', 'doll', 'dream', 'day', 'date']})
print(df_20)
```

```python
df_20.pivot(index = 'A', columns = 'B', values = 'C')
```

```python
df_20.pivot(index = 'Sr no.', columns = 'A')['B']
```

```python
df_20.pivot(index = 'Sr no.', columns = 'A', values = ['B', 'C'])
```
