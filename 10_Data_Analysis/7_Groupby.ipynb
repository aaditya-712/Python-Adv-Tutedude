```python
# GROUPBY
import numpy as np
grp = df.groupby("A")
grp.aggregate(np.sum)
```

```python
grp = df.groupby(["A","B"])
grp.aggregate(np.sum)
```

```python
grp = df.groupby(["A","B"], as_index = False)
grp.aggregate(np.sum)
```
